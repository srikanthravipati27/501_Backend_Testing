```python
import pandas as pd
from pandasql import sqldf
fun=lambda q:sqldf(q,globals())
customers=pd.read_csv('customers.csv')
purchases=pd.read_csv('purchases.csv')
```

```python
customers.head()
```

```python
purchases.head()
```

```python
que1="select * from customers join purchases on customers.id=purchases.customer_id"
res=fun(que1)
print(res)
```

```python
#group the result by custer_id,name ,age
que2="select customer_id,name,age from res group by customer_id,name ,age "
grouped=fun(que2)
print(grouped)
```

```python
que3="select * from customers c join purchases p on c.id=p.customer_id where c.age<30 and p.quantity>5"
res3=fun(que3)
print(res3)
```

```python
que4="select p.customer_id,count(distinct p.product_id) as co from customers c join purchases p on c.id=p.customer_id group by c.id having co>1  "
res4=fun(que4)
print(res4)
```

```python
que5="select c.id,sum(p.quantity) from customers c join purchases p on c.id=p.customer_id group by c.id "
res5=fun(que5)
print(res5)
```
